1.Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")
words = sentence.split()        
unique_words = set(words)       
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

2.Highest Salary from Employee Data

```python
employees = {
    "Reena": 80000,
    "Priya": 75000,
    "kiran": 65000
}
highest_employee = max(employees, key=employees.get)
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

3.Find Maximum and Minimum Values

```python
numbers = [46, 24, 88, 20, 65]
max_value = numbers[0]
min_value = numbers[0]
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

4.Count Products Above a Price Threshold

```python
prices = [460, 1400, 900, 1800, 200]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("product above 1000:", count)        
```

5.Calculate Attendance Percentage

```python
attendance = ["T", "A", "M", "I", "L"]
present_count = 0
for day in attendance:
    if day == "T":
        present_count += 1
total_days = len(attendance)
percentage = (present_count / total_days) * 100
print("Attendance Percentage:", percentage)
```

6.Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543256, 2345678954, 9912345678]
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

7.Count Character Frequency

```python
text = "python"
char_count = {}
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1
print(char_count)
```

10.Calculate Average Marks

```python
marks = [70, 75, 80, 65]   
total = sum(marks)
average = total / len(marks)
print("Average Marks:", average)
```
